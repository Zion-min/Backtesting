```python
#위에과정을 함수로 만들기 
#왜 why? 리밸런싱마다 유니버스 추출 과정 반복 
#함수 각 과정 colab -> 함수 순서 기재해놓음 참고
#시가총액 200위 인종목들 뽑아내기 -> 유니버스 안의 pbr, 가격모멘텀 뽑아내는데 쓸거에요
def get_universe_by_cap(df, start_date, cap):
    close=df
    close=close.loc[start_date:]
    close=close.transpose()
    close=close[close.index.isin(cap)][:200]
    close=close.transpose()
    return close
```

```python
#함수: pbr, 3개월 가격모멘텀을 200개 시가총액 순위 & 시작날짜에 맞춰주는 함수
#이제 따라치시면 돼요!
def set_universe(df, pbr, start_date):
    cap = stock.get_market_cap_by_ticker(start_date)
    cap = cap.sort_values(by='시가총액', ascending=False)
    cap = cap[cap.index.isin(pbr.transpose().index)].index
    
    close = get_universe_by_cap(df.fillna(0),  start_date-dt.timedelta(days=90),  cap)
    price_mom = close.pct_change(60).loc[start_date:]
    
    pbr = get_universe_by_cap(pbr, start_date-dt.timedelta(days=90), cap)
    pbr = pbr.loc[start_date:]
    return price_mom, pbr
```

```python
def get_portpolio(pbr, mom, start_date):
    portpolio = pd.DataFrame({'PBR':pbr.loc[start_date], '3MOM':mom.loc[start_date]})
    #pbr 상위 10, mom 상위 10
    #종목들 
    mixed_pp = portpolio.sort_values(by='PBR')[:10].index
    mixed_pp = mixed_pp.append( portpolio.sort_values(by='3MOM', ascending=False)[:10].index )
    
    #integ 혼합해서 랭크매기고, 20개 뽑기
    portpolio['PBR_rank'] = portpolio['PBR'].rank(method='min', ascending=True)
    portpolio['3MOM_rank'] = portpolio['3MOM'].rank(method='min', ascending=False)
    portpolio['total_rank'] = portpolio['PBR_rank']+portpolio['3MOM_rank']
    integrated_pp = portpolio.sort_values(by='total_rank', ascending=True)[:20].index  
    return mixed_pp, integrated_pp
```

```python
def get_returns(close, mixed_pp, integrated_pp, start_date, end_date, mixed_cagr, integ_cagr):
    #mix, integ 누적수익률
    mix_returns = ((close[mixed_pp]/close[mixed_pp].iloc[0]).sum(axis=1)/20).loc[start_date:end_date]*mixed_cagr
    integ_returns = ((close[integrated_pp]/close[integrated_pp].iloc[0]).sum(axis=1)/20).loc[start_date:end_date]*integ_cagr
    return mix_returns, integ_returns
    
```

```python
def DD(returns):
    전고점 = returns.cummax()
    DD = (returns/전고점)-1
    return DD

def bench_DD(benchmark):
    cum = (benchmark+1).cumprod()
    전고점 = cum.cummax()
    DD = (cum/전고점)-1
    return DD

def MDD(returns):
    return np.min(DD(returns))


def CAGR(returns):
    return (1+returns).prod()-1

def information_ratio(returns, benchmark):
    return np.mean(returns - benchmark)/np.std(returns - benchmark)
```
