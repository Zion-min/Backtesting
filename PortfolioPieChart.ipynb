```python
#Description: This program visualizes my stock portfolio
```

```python
#import the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
#data
df = pd.DataFrame(columns=['Ticker', 'Amount', 'Close'])
df['Ticker'] = ['FB', 'AAPL', 'AMZN', 'NFLX', 'GOOG']
df['Amount'] = [10, 90, 3, 50, 1]
df['Close'] = [132.64, 123.00, 3161.12, 539.42, 2317.50]
```

```python
#get the total invested amount of money in the portfolio
Portfolio_Total_Amount = sum( df['Amount'] * df['Close'])
Portfolio_Total_Amount = round(Portfolio_Total_Amount, 2)
Portfolio_Total_Amount
```

```python
#Visually show the portfolio and some additional information
stock_tickers = df['Ticker'].values
sizes = df['Amount'] * df['Close']

#5개의 행들만큼의 0 리스트를 만들어낸다.
#n은 0~4까지 무작위로 지정된 숫자임. listOfZeros 리스트에 무작위로 0.1을 넣는다 
#아 이거 때문에 삐져나온 모양 결정되는 거임(explode)
listOfZeros = [0] * df.shape[0]
n = random.randint(0, df.shape[0] - 1)
listOfZeros[n] = 0.1
explode = listOfZeros

#Create a figure
#넓이 지정
fig1, ax1 = plt.subplots(figsize = (10,10))
#plot the pie chart
ax1.pie(sizes, explode=explode, labels=stock_tickers, autopct='%.2f%%',shadow='True', startangle=360)

#set the title
ax1.set_title('Portfolio Pie Chart', fontsize=22)

#Add text to the visual
x = -1.75
y = 1
ax1.text(x,y, 'Overview:', fontsize=24, color='RoyalBlue')

y_counter = 0.12
ax1.text(x,y-y_counter, 'Total: '+str(Portfolio_Total_Amount), fontsize=15, color='SkyBlue')
for i in range(0, df.shape[0]):
    ax1.text(x, 0.88 - y_counter, df['Ticker'][i] + ': $' + str(round(df['Amount'][i] * df['Close'][i], 2)), fontsize=14, color='Gray')
    y_counter = y_counter + 0.12
```
