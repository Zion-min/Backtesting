```python
import pandas as pd
import numpy as np
import requests
```

```python
df = pd.read_excel('2000_2021_close.xlsx')
```

```python
df.set_index(pd.DatetimeIndex(df['Code'].values), inplace=True)
```

```python
df.drop(columns=['Code'], axis=1, inplace=True)
```

```python
df1 = df.dropna(axis=1)
df1
```

```python
assets = df1.columns
```

```python
from pypfopt.efficient_frontier import EfficientFrontier
from pypfopt import risk_models
from pypfopt import expected_returns
```

```python
mu = expected_returns.mean_historical_return(df1)
S = risk_models.sample_cov(df1)
```

```python
ef = EfficientFrontier(mu, S)
```

```python
weights = ef.max_sharpe()
```

```python
cleand_weights = ef.clean_weights()
```

```python
ef.portfolio_performance(verbose=True)
```

```python
#Get the discrete allocation of each share per stock
import cvxpy as cp
from pypfopt.discrete_allocation import DiscreteAllocation, get_latest_prices


portfolio_val = 5_000_000
latest_prices = get_latest_prices(df1)
weights = cleand_weights
da = DiscreteAllocation(weights, latest_prices, total_portfolio_value=portfolio_val)
allocation, leftover = da.lp_portfolio()
print('Discrete allocation:', allocation)
print('Funds Remaining: \\', leftover)
```
